## Using the matplotlib library to look at a picture as numerical data.

```python
from matplotlib import pyplot
test_picture = pyplot.imread("picture1.jpg")
print("Numpy array of the image is: ", test_picture)
pyplot.imshow(test_picture)
```

```python
test_picture_filtered = 2*test_picture/3
pyplot.imshow(test_picture_filtered)
```

Displayed I see the same graph as the previous command, however the image seems to be scaled down to its RGB colours. I think that the first command (test_picture_filtered = 2*test_picture/3) changes the brightness of the image and scales down the pixels. The second is what displays the modified image.
